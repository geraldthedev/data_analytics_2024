```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk


movie_data_url ='https://raw.githubusercontent.com/fivethirtyeight/data/refs/heads/master/fandango/fandango_score_comparison.csv'
movie_scrape_url = 'https://raw.githubusercontent.com/fivethirtyeight/data/refs/heads/master/fandango/fandango_scrape.csv'

movie_pd = pd.read_csv(movie_data_url)
movie_scrapes_pd = pd.read_csv(movie_scrape_url)

movie_scores = pd.DataFrame(movie_pd[['FILM','RottenTomatoes','Metacritic','Metacritic_User','Metacritic_user_vote_count','IMDB_user_vote_count', 'IMDB']])
movie_scrapes_scores = pd.DataFrame(movie_scrapes_pd[['FILM','RATING','VOTES']])

joined = movie_scores.merge(movie_scrapes_scores, how='inner', on='FILM')

x = 100
y = 5.0
z = 10.0

x_vals = joined[['RottenTomatoes','Metacritic']].values
y_vals = joined[['RATING']].values
z_vals = joined[['Metacritic_User']].values

fig = plt.figure(figsize=(5,5))
ax= fig.add_subplot(111, projection='3d')
ax.scatter(x_vals,y_vals,z_vals)



plt.show()
```
